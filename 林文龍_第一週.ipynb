## 挑戰題一：買相機
題目說明：財團法人大兔兔文教基金會要添購攝影設備。
他們預計在未來四個月內分別添購四項價格為 100k,150k,230k,80k的設備，
請以折線圖畫出累積消費額，並以圓餅圖畫出各器材所佔比例。

```python
a = int(100)
b = int(150)
c = int(230)
d = int(80)

expend = [a, a+b, a+b+c, a+b+c+d]
plot(expend)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(expend)
plt.show()
```
